```python
import numpy as np
import pandas as pd
import cv2
from xml.dom.minidom import parse
import xml.dom.minidom
```

```python
pathOfXML='interview/task2/dataset/annotations.xml'
DOMTree = xml.dom.minidom.parse(pathOfXML)
collection = DOMTree.documentElement
annotation=DOMTree.getElementsByTagName("annotations")[0]
images=annotation.getElementsByTagName("image")
iamgeDetails=[]

for image in images:
    
    filename=image.getAttribute("id")
    completeName=filename+".jpg"
    boxes=image.getElementsByTagName("box")
    for box in boxes:
        headDetail=[]
        boxLabel=box.getAttribute("label")
        
        if(boxLabel == "head"):
            has_safety_helmet="no"
            mask="no"
            xtl=box.getAttribute("xtl")
            ytl=box.getAttribute("ytl")
            xbr=box.getAttribute("xbr")
            ybr=box.getAttribute("ybr")
            attributes=box.getElementsByTagName("attribute")
            for attribute in attributes:
            
                if(attribute.getAttribute("name")=="has_safety_helmet"):
                    
                    if(str(attribute.childNodes[0].data)=="yes"):
                        
                        has_safety_helmet="yes"
            
                if(attribute.getAttribute("name")=="mask"):
                    
                    if(str(attribute.childNodes[0].data)=="yes"):
        
                        mask="yes"
                
            headDetail.append(filename)
            headDetail.append(xtl)
            headDetail.append(ytl)
            headDetail.append(xbr)
            headDetail.append(ybr)
            headDetail.append(has_safety_helmet)
            headDetail.append(mask)
            iamgeDetails.append(headDetail)
            
print("imageName,xtl,ytl,xbr,ybr,has_safety_helmet,has_mask")            
for i in range(len(iamgeDetails)):
    print(iamgeDetails[i])
            
            
            
            
            
            
    
    
    
```

```python
#croping and saving the images
#using 4 class 
#class 1 if both mask and helmet are present
#class 2 if only helmet are present
#class 3 if only mask are present
#class 4 if both mask and helmet are not present
for i in range(len(iamgeDetails)):
    imgDetail=iamgeDetails[i]
    img = cv2.imread('interview/task2/dataset/images/'+imgDetail[0]+'.jpg')
    #cv2.imshow("image",img)
    #cv2.waitKey(1)
    y1=int(float(imgDetail[2]))
    y2=int(float(imgDetail[4]))
    x1=int(float(imgDetail[1]))
    x2=int(float(imgDetail[3]))
    cropImg=img[y1:y2,x1:x2]
    newName=""
    if(imgDetail[5]== "yes" and imgDetail[6]== "yes"):
        newName=imgDetail[0]+'_1.jpg'
        newFullName='interview/task2/dataset/imagesCrop/1/'+newName
    if(imgDetail[5]== "yes" and imgDetail[6]== "no"):
        newName=imgDetail[0]+'_2.jpg'
        newFullName='interview/task2/dataset/imagesCrop/2/'+newName
    if(imgDetail[5]== "no" and imgDetail[6]== "yes"):
        newName=imgDetail[0]+'_3.jpg'
        newFullName='interview/task2/dataset/imagesCrop/3/'+newName
    if(imgDetail[5]== "no" and imgDetail[6]== "no"):
        newName=imgDetail[0]+'_4.jpg'
        newFullName='interview/task2/dataset/imagesCrop/4/'+newName
    
    #print(newFullName)
    cv2.imwrite(newFullName,cropImg)
    cv2.waitKey(1)
    
```

```python
# Importing all necessary libraries
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
```

```python
img_width, img_height = 224, 224
train_data_dir = 'interview/task2/dataset/imagesCrop/Train/'
validation_data_dir = 'interview/task2/dataset/imagesCrop/Test/'
nb_train_samples =693 
nb_validation_samples = 176
epochs = 10
batch_size = 16
```

```python
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
  
model = Sequential()
model.add(Conv2D(32, (2, 2), input_shape = input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2, 2)))
  
model.add(Conv2D(32, (2, 2)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2, 2)))
  
model.add(Conv2D(64, (2, 2)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size =(2, 2)))
  
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
  
model.compile(loss ='binary_crossentropy',
                     optimizer ='rmsprop',
                   metrics =['accuracy'])
  
train_datagen = ImageDataGenerator(
                rescale = 1. / 255,
                 shear_range = 0.2,
                  zoom_range = 0.2,
            horizontal_flip = True)
  
test_datagen = ImageDataGenerator(rescale = 1. / 255)
  
train_generator = train_datagen.flow_from_directory(train_data_dir,
                              target_size =(img_width, img_height),
                     batch_size = batch_size, class_mode ='binary')
  
validation_generator = test_datagen.flow_from_directory(
                                    validation_data_dir,
                   target_size =(img_width, img_height),
          batch_size = batch_size, class_mode ='binary')
  
model.fit_generator(train_generator,
    steps_per_epoch = nb_train_samples // batch_size,
    epochs = epochs, validation_data = validation_generator,
    validation_steps = nb_validation_samples // batch_size)
  
model.save_weights('model_saved.h5')
```
