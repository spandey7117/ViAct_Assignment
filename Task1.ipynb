```python
#Question 2 to search all parents
def findAllParent(labelName, parentLabelList,childLabelList):
    allParents=[]
    indices = [i for i, x in enumerate(childLabelList) if x == labelName]
    for i in range(len(indices)):
            allParents.append(parentLabelList[indices[i]])
            #print(parentLabelList[indices[i]])
    return allParents;
```

```python
#Question One find all Sibling
def findSibling(labelName, parentLabelList,childLabelList):
    sibling=[]
    allParent=findAllParent(labelName, parentLabelList,childLabelList)
    for i in range(len(allParent)):
        parentNameOfLabel=allParent[i]
        #print(parentNameOfLabel)
        for j in range(len(parentLabelList)):
            if(parentLabelList[j] is parentNameOfLabel):
                sibling.append(childLabelList[j])
                #print(childLabelList[j])
            
    return sibling
```

```python
def findAllAncestor(labelName, parentLabelList,childLabelList,allAncestor):
    allParent=findAllParent(labelName,parentLabelList,childLabelList)
    for i in range(len(allParent)):
        #print(allParent[i])
        if(allParent[i] not in allAncestor):
            allAncestor.append(allParent[i])
            findAllAncestor(allParent[i], parentLabelList,childLabelList,allAncestor)
    return allAncestor
```

```python
def findIfFromSameAncestor(labelName1,labelName2, parentLabelList,childLabelList):
    allAncestor1=[]
    allAncestor2=[]
    allAncestor1=findAllAncestor(labelName1, parentLabel,childLabel,allAncestor1)
    allAncestor2=findAllAncestor(labelName2, parentLabel,childLabel,allAncestor2)
    #print(allAncestor1)
    #print(allAncestor2)
    for i in range(len(allAncestor1)-1):
        for j in range(len(allAncestor2)-1):
                if(allAncestor1[i] is allAncestor2[j]):
                    return True
    return False
```
